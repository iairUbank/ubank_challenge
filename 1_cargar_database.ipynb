# Cargo base de datos en PostgreSQL 
Para realizar esta tarea desde pgAdmin 4, se creó una base de datos denominada 'Ubank'. Inmediatamente se crearon las 4 tablas correspondientes desde Query Tool:

**catalog**

CREATE TABLE public.catalog

(

    project_id VARCHAR(50) PRIMARY KEY,
	project_name VARCHAR(50),
	project_type VARCHAR(50)
);


**test_projects:** 

CREATE TABLE public.test_projects

(

    project_id VARCHAR(50),
	project_name VARCHAR(50),
	goal_date TIMESTAMP,
	user_id VARCHAR(50),
	project_category_id VARCHAR(50),
	total NUMERIC
);


**test_rules:** 

CREATE TABLE public.test_rules

(

	rule_id VARCHAR(50),
	project_id VARCHAR(50),
	rule_type VARCHAR(50),
	amount NUMERIC,
	frecuency SMALLINT,
	categories VARCHAR(50)	
);


**test_transactions:** 

CREATE TABLE public.test_transactions

(

	user_id VARCHAR(50),
	description VARCHAR(50),
	transaction_date TIMESTAMP,
	amount NUMERIC
);

Observar que defino una columna como PRIMARY KEY en la tabla 'catalog' y podría hacerlo para otras columnas de las demás tablas, así como también podría definir algunas FOREING KEY's. Sin embargo las tablas y columnas no son tan numerosas y puedo distinguirlas sin dichas etiquetas a la hora de una eventual unión entre tablas, por lo que avanzo sin detenerme mucho en esto.

Finalmente importo los archivos .csv a las tablas creadas y comienzo a realizar consultas (ir a file '2_consultas').
